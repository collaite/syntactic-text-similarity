```python
import pandas as pd

df = pd.read_csv('data/results/syntactic/results-syn-hamming.csv')
```

```python
# dataframe with context information excluding the 'length' measure
df = df.loc[(~df['similarity_metric'].isin(['length', 'lcsseq', 'mlipns', 'prefix', 'postfix', 'identity', 'matrix']))]
# [df['similarity_metric'] != 'length']
# dataframe with context information only for "vrouwenhand" token
df_vrouwenhand = df[df['source_token'] == 'vrouwenhand']
# dataframe with information only for exact matches for contexts
df_exact_matches = df[df['similarity_score'] == 1.0]
# dataframe with information only for nonexact matches for contexts
df_nonexact_matches = df[df['similarity_score'] < 1.0]
# dataframe with information only for context matching scores of 0.0
df_completely_unrelated = df[df['similarity_score'] == 0.0]
what = df[df['similarity_score'] < 0.0]
# dataframe with information only for context matching scores greater than 0.0
df_related = df[df['similarity_score'] > 0.0]
```

```python
s = set(df_nonexact_matches['source_token'].unique())
diff = [x for x in df_exact_matches['source_token'].unique() if x not in s]
print('Stable tokens: ', diff)
print()
print("A 'stable' token is one that perfectly matches across all witnesses (shares same context)")
print("No. of stable tokens: ", len(diff))
print()
print('Non-stable tokens: ', df_nonexact_matches['source_token'].unique())
print("A 'non-stable' token is one that is not a stable token")
print("No. of non-stable tokens: ", str(df['source_token'].nunique() - len(diff)))
print()
s2 = set(df_exact_matches['source_token'].unique())
diff2 = [x for x in df_nonexact_matches['source_token'].unique() if x not in s2]
print('Unstable tokens: ', diff2)
print("An 'unstable' token is one that never matches perfectly with any witness using any measure")
print("No. of unstable tokens: ", len(diff2))
print()

# dataframe with context information only for stable tokens
exact_matches_perfect_scores = df.loc[(df['source_token'].isin(diff))]
# dataframe with context information only for non-stable tokens
exact_matches_imperfect_scores = df.loc[(~df['source_token'].isin(diff))]
```
